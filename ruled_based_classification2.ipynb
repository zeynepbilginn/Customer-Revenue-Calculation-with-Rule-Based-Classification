```python
agg_df = df.groupby(["COUNTRY","SOURCE","SEX","AGE"]).agg({"PRICE": "mean"}).sort_values("PRICE",ascending = False)
agg_df
```

```python
agg_df.reset_index(inplace = True)
```

```python
def change_dtype_age(x):
    if x < 18:
        return "0_18"
    elif x < 23:
        return "19_23"
    elif x < 30:
        return "24_30"
    elif x < 40:
        return "31_40"
    else:
        return "41_70"
    
agg_df["AGE_CAT"] = agg_df["PRICE"].apply(lambda x: change_dtype_age(x))
```

```python
agg_df
```

```python
agg_df["customer_level_based"] = agg_df["COUNTRY"].str.upper() + "_" + agg_df["SOURCE"].str.upper() + "_" + agg_df["SEX"].str.upper() + "_" + agg_df["AGE_CAT"].str.upper()
```

```python
agg_df
```

```python
customer_level_based_df = agg_df.loc[:,["customer_level_based","PRICE"]]
```

```python
customer_level_based_df
```

```python
customer_level_based_df["SEGMENT"] = pd.qcut(customer_level_based_df["PRICE"],4, labels = ["D","C","B","A"])
```
