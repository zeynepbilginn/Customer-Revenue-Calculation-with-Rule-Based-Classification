SaleCheckInDayDiff değişkenini kategorik bir değişkene çevirelim.


```python
def change_dtype_day_diff(day_diff):
    if day_diff < 7:
        return 'Last Minuters'
    elif day_diff < 30:
        return "Potential Planners"
    elif day_diff < 90:
        return "Planners"
    else:
        return  'Early Bookers'

df["EB_Score"] = df["SaleCheckInDayDiff"].apply(lambda x: change_dtype_day_diff(x))
```
